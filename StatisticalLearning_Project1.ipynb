```python
import numpy as np
import pandas as pd
```

```python
#Reading the source data
df= pd.read_csv("C:\\Users\\jaysh\\OneDrive\\Documents\\Project\\Project1- Statistics\\Laliga.csv")
```

```python
#Assigning the first row as column header
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
```

```python
df.columns
```

```python
#Replace dashes with O from the dataset
df= df.replace("-",0)
```

```python
#Fetching all the teams which started playing between 1930-1980
df1 = df.loc[df["Debut"]>"1930<1980",["Team"]]
df1
```

```python
#Converting data type for particular columns to integer to perform arithmetic opertaions
datatype_convert = {'Seasons':int, 'Points':int,'GamesPlayed':int,'GamesWon':int,'GamesDrawn':int,'GamesLost':int,'GoalsFor':int,
                  'GoalsAgainst':int, 'Champion':int,'Runner-up':int,'Third':int,'Fourth':int,'Fifth':int,'Sixth':int,'T':int
                   }
df = df.astype(datatype_convert)
```

```python
#Finding out the datatype of all the columns in the dataframe
df.dtypes
```

```python
#Top 5 Teams by Points scored
Top_Teams = df.nlargest(5,'Points')
Top_Teams[["Team","Points"]]
```

```python
# Creating a customised function to calculate the goal difference
def Goal_diff_count(x,y):
    return x-y
```

```python
#Goal_diff is imputed in the dataframe for all the rows using the function defined
df["Goal_diff_count"] = Goal_diff_count(df["GoalsFor"],df["GoalsAgainst"])
```

```python
# Printing the team name with maximum goal difference
print(df[["Team","Goal_diff_count"]].max())
print(df[["Team","Goal_diff_count"]].min())
```

```python
df["Winning_percent"] = df["GamesWon"]/df["GamesPlayed"]*100
df.head(5)
```

```python
# Checking whether there are Nan values in Winning_percent column
check_nan = np.isnan(df["Winning_percent"])
check_nan[df["Team"]].index
```

```python
#Replace all na values from winning_percent column to 0
df["Winning_percent"].fillna("0")
```

```python
#Top 5 teams with highest winning percentage
Top_5 = df.nlargest(5,"Winning_percent")
print(Top_5["Team"])
```

```python
#Group by Best position and sumup the points scored
df_bestposition = df.groupby("BestPosition")["Points"].agg("sum")
df_bestposition
```

```python
#Group by bestposition and team and sum of points scored by them
df_bestposition_1 = df.groupby(by=["BestPosition","Team"])["Points"].agg("sum")
df_bestposition_1
```
